```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan state/kebiasaan
states = ["Tidak Menggunakan", "Menggunakan Sesekali", "Menggunakan Sering"]

# Indeks masing-masing state
state_to_index = {state: i for i, state in enumerate(states)}

# Matriks transisi, misalnya:
transition_matrix = np.array([
    [0.6, 0.3, 0.1],  # Dari "Tidak Menggunakan"
    [0.2, 0.5, 0.3],  # Dari "Menggunakan Sesekali"
    [0.1, 0.4, 0.5]   # Dari "Menggunakan Sering"
])

# Fungsi untuk mensimulasikan prediksi penggunaan media sosial dengan Markov Chain
def simulate_markov_chain(transition_matrix, initial_state, steps):
    current_state = initial_state
    history = [current_state]

    for _ in range(steps):
        # Probabilitas transisi berdasarkan current state
        current_state = np.random.choice(
            range(len(states)),
            p=transition_matrix[current_state]
        )
        history.append(current_state)

    return history

# Inisialisasi state awal
initial_state = state_to_index["Tidak Menggunakan"]

# Simulasikan prediksi untuk 30 hari ke depan
steps = 30
history = simulate_markov_chain(transition_matrix, initial_state, steps)

# Konversi hasil simulasi ke dalam nama state
history_states = [states[state] for state in history]

# Visualisasi hasil
plt.figure(figsize=(12, 6))
plt.plot(range(steps + 1), history, marker='o', color='b', linestyle='-')
plt.xticks(range(steps + 1))
plt.yticks(range(len(states)), states)
plt.xlabel("Hari")
plt.ylabel("Kebiasaan Penggunaan Media Sosial")
plt.title("Simulasi Prediksi Kebiasaan Penggunaan Media Sosial dengan Markov Chain")
plt.grid(True)
plt.show()

# Print hasil simulasi
print("Hasil Simulasi:")
for day, state in enumerate(history_states):
    print(f"Hari {day + 1}: {state}")
```
